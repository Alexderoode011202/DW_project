# Some plots of the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file = '../new-data/everything_merged_dataset_no_NULL_violent&sexual.csv'
df = pd.read_csv(file)
```

Distribution plots for the types of weapons, crimes, and GDP

```python
sns.set(style="whitegrid")

firearm_columns = [
    "Estimate of firearms in civilian possession per 1,000",
    "Total law enforcement firearms per 1,000",
    "Total military firearms per 1,000"
]

plt.figure(figsize=(12, 6))
for column in firearm_columns:
    sns.kdeplot(df[column], fill=True, label=column.split(" per")[0])

plt.title("Distribution of Firearm Ownership per 1,000 People", fontsize=16)
plt.xlabel("Firearms per 1,000 People", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend(title="Firearm Ownership Type", fontsize=10)
plt.show()
```

```python
crime_columns = [
    "Total Homicides per 1,000",
    "Total Sexual Crimes per 1,000",
    "Total Violent Crimes per 1,000"
]

plt.figure(figsize=(12, 6))
for column in crime_columns:
    sns.kdeplot(df[column], fill=True, label=column.split(" per")[0])

plt.title("Distribution of Crime Rates per 1,000 People", fontsize=16)
plt.xlabel("Crime Rate per 1,000 People", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend(title="Crime Type", fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df["Country GDP per 1,000"], fill=True, color="purple")
plt.title("Distribution of GDP per Capita", fontsize=16)
plt.xlabel("GDP per 1,000 People (in USD)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.show()
```

Proportion of types of weapons per country. The plot looks cluttere because there a lot of countries.

```python
columns_to_plot = [
    "Estimate of firearms in civilian possession per 1,000",
    "Total law enforcement firearms per 1,000",
    "Total military firearms per 1,000"
]

country_firearm_data = df[["ISO"] + columns_to_plot]

country_firearm_data = country_firearm_data.set_index("ISO")

country_firearm_data.plot(kind="bar", figsize=(20, 8), width=0.8)
plt.title("Firearm Ownership per 1,000 People by Country", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Firearms per 1,000 People", fontsize=12)
plt.legend(title="Firearm Ownership Type", fontsize=10)
plt.xticks(rotation=45, fontsize=10)  
plt.tight_layout()
plt.show()
```

Plot the relationship betweenthe types of weapon ownership and the different crime rates (+ combined crime rate). *Note*: not all combinations are visualized because it will get too long, but they can be added further if needed.

```python
# Civil vs homicide
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Estimate of firearms in civilian possession per 1,000",
    y="Total Homicides per 1,000",
    data=df,
    hue="Continent",  
    palette="tab10",
    s=100
)
plt.title("Civilian Firearms vs. Homicides per 1,000 People", fontsize=16)
plt.xlabel("Civilian Firearms per 1,000 People", fontsize=12)
plt.ylabel("Homicides per 1,000 People", fontsize=12)
plt.legend(title="Continent")
plt.show()

# Law enforcement vs. sexual crimes
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Total law enforcement firearms per 1,000",
    y="Total Sexual Crimes per 1,000",
    data=df,
    hue="Continent",
    palette="tab10",
    s=100
)
plt.title("Law Enforcement Firearms vs. Sexual Crimes per 1,000 People", fontsize=16)
plt.xlabel("Law Enforcement Firearms per 1,000 People", fontsize=12)
plt.ylabel("Sexual Crimes per 1,000 People", fontsize=12)
plt.legend(title="Continent")
plt.show()

# Military firearms vs. violent crimes
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Total military firearms per 1,000",
    y="Total Violent Crimes per 1,000",
    data=df,
    hue="Continent",
    palette="tab10",
    s=100
)
plt.title("Military Firearms vs. Violent Crimes per 1,000 People", fontsize=16)
plt.xlabel("Military Firearms per 1,000 People", fontsize=12)
plt.ylabel("Violent Crimes per 1,000 People", fontsize=12)
plt.legend(title="Continent")
plt.show()

# Firearms vs. overall crime Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Estimate of firearms in civilian possession per 1,000",
    y="Combined Crime Rate per 1,000",
    data=df,
    hue="Continent",
    palette="tab10",
    s=100
)
plt.title("Civilian Firearms vs. Overall Crime Rate per 1,000 People", fontsize=16)
plt.xlabel("Civilian Firearms per 1,000 People", fontsize=12)
plt.ylabel("Combined Crime Rate per 1,000 People", fontsize=12)
plt.legend(title="Continent")
plt.show()
```

Plot the relationship between GDP and different types of crime (+ overall rate)

```python
# GDP per 1,000 vs. homicides
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Country GDP per 1,000",
    y="Total Homicides per 1,000",
    data=df,
    hue="Continent",
    palette="Set2",
    s=100
)
plt.title("GDP per 1,000 vs. Homicides per 1,000 People", fontsize=16)
plt.xlabel("GDP per 1,000 People (in USD)", fontsize=12)
plt.ylabel("Homicides per 1,000 People", fontsize=12)
plt.legend(title="Continent")
plt.show()

# GDP per 1,000 vs. violent crimes
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Country GDP per 1,000",
    y="Total Violent Crimes per 1,000",
    data=df,
    hue="Continent",
    palette="Set2",
    s=100
)
plt.title("GDP per 1,000 vs. Violent Crimes per 1,000 People", fontsize=16)
plt.xlabel("GDP per 1,000 People (in USD)", fontsize=12)
plt.ylabel("Violent Crimes per 1,000 People", fontsize=12)
plt.legend(title="Continent")
plt.show()

# GDP per 1,000 vs. vverall crime rate
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="Country GDP per 1,000",
    y="Combined Crime Rate per 1,000",
    data=df,
    hue="Continent",
    palette="Set2",
    s=100
)
plt.title("GDP per 1,000 vs. Overall Crime Rate per 1,000 People", fontsize=16)
plt.xlabel("GDP per 1,000 People (in USD)", fontsize=12)
plt.ylabel("Combined Crime Rate per 1,000 People", fontsize=12)
plt.legend(title="Continent")
plt.show()
```

Heatmap for correlation between all variables

```python
correlation_columns = [
    "Estimate of firearms in civilian possession per 1,000",
    "Total law enforcement firearms per 1,000",
    "Total military firearms per 1,000",
    "Total Homicides per 1,000",
    "Total Sexual Crimes per 1,000",
    "Total Violent Crimes per 1,000",
    "Combined Crime Rate per 1,000",  
    "Country GDP per 1,000"
]

correlation_matrix = df[correlation_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix,
    annot=True, 
    cmap="coolwarm",  
    fmt=".2f",  
    linewidths=0.5,  
    cbar_kws={"shrink": 0.8} 
)
plt.title("Correlation Heatmap: Firearms, Crime Rates, and GDP", fontsize=16)
plt.show()
```
