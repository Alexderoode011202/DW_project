# Further cleaning the datasets and creating a combined dataset

New section added below

```python
import pandas as pd
```

```python
df1 = pd.read_csv('./separate_datasets/civilian dataset.csv')
df1.rename(columns={'Population 2017': 'Population'}, inplace=True)
df1.rename(columns={'Country code': "ISO"}, inplace=True)
df1.rename(columns={'Region': "Continent"}, inplace=True)
df1.head(10)
```

```python
df2 = pd.read_csv('./separate_datasets/law_enforcement dataset.csv')
df2.head(10)
```

```python
df3 = pd.read_csv('./separate_datasets/military dataset.csv')
df3.rename(columns={'Country Code': 'ISO'}, inplace=True)
df3.rename(columns={'Region': "Continent"}, inplace=True)
df3.head(10)
```

```python
# Delete some redundant rows from the dataframes
value_to_remove = "ISO"
df1 = df1[df1["ISO"] != value_to_remove]
df2 = df2[df2["ISO"] != value_to_remove]
df3 = df3[df3["ISO"] != value_to_remove]

value_to_remove1 = "Country"
df1 = df1[df1["ISO"] != value_to_remove1]
df2 = df2[df2["ISO"] != value_to_remove1]
df3 = df3[df3["ISO"] != value_to_remove1]
```

```python
# With country name
set1 = set(df1['Country'])
set2 = set(df2['Country'])
set3 = set(df3['Country'])

if set1 == set2 == set3:
    print("All DFs have the same countries!!!!")
else: 
    missing_in_df2_and_df3 = set1 - set2 - set3
    missing_in_df1_and_df3 = set2 - set1 - set3
    missing_in_df1_and_df2 = set3 - set1 - set2

    diff_df1_df2 = set1.symmetric_difference(set2)
    diff_df1_df3 = set1.symmetric_difference(set3)
    diff_df2_df3 = set2.symmetric_difference(set3)

    all_countries = set1.union(set2, set3)

    missing_in_df1 = all_countries - set1
    missing_in_df2 = all_countries - set2
    missing_in_df3 = all_countries - set3

    print(f"Num missing df1 - {len(missing_in_df1)}: {missing_in_df1}")
    print(f"Num missing df2 - {len(missing_in_df2)}: {missing_in_df2}")
    print(f"Num missing df3 - {len(missing_in_df3)}: {missing_in_df3}")
```

```python
# With country code (ISO)
set1 = set(df1['ISO'])
set2 = set(df2['ISO'])
set3 = set(df3['ISO'])

if set1 == set2 == set3:
    print("All DFs have the same countries!!!!")
else: 
    missing_in_df2_and_df3 = set1 - set2 - set3
    missing_in_df1_and_df3 = set2 - set1 - set3
    missing_in_df1_and_df2 = set3 - set1 - set2

    diff_df1_df2 = set1.symmetric_difference(set2)
    diff_df1_df3 = set1.symmetric_difference(set3)
    diff_df2_df3 = set2.symmetric_difference(set3)

    all_countries = set1.union(set2, set3)

    missing_in_df1 = all_countries - set1
    missing_in_df2 = all_countries - set2
    missing_in_df3 = all_countries - set3

    print(f"Num missing df1 - {len(missing_in_df1)}: {missing_in_df1}")
    print(f"Num missing df2 - {len(missing_in_df2)}: {missing_in_df2}")
    print(f"Num missing df3 - {len(missing_in_df3)}: {missing_in_df3} \n")

    combined = list(missing_in_df1) + list(missing_in_df2) + list(missing_in_df3)
    combined = set(combined)

    print(f'Total mismatching ISO codes ({len(combined)}): {combined}')
```

```python
# Create a merged dataset with only the essential info
for df in [df1, df2, df3]:
    df["Population"] = pd.to_numeric(df["Population"], errors="coerce")

merged_df = pd.merge(df1, df2, on=["ISO", "Country", "Continent", "Subregion", "Population"], how="outer")
merged_df = pd.merge(merged_df, df3, on=["ISO", "Country", "Continent", "Subregion", "Population"], how="outer")

final_columns = [ "ISO", "Country", "Continent", "Subregion", "Population"]

merged_df = merged_df[final_columns]

# Mappings from df1 to merged_df
m1 = df1.set_index("ISO")["Estimate of firearms in civilian possession"]
merged_df["Estimate of firearms in civilian possession"] = merged_df["ISO"].map(m1)
m2 = df1.set_index("ISO")["Registered firearms"]
merged_df["Registered firearms"] = merged_df["ISO"].map(m2)
m3 = df1.set_index("ISO")["Unregistered firearms"]
merged_df["Unregistered firearms"] = merged_df["ISO"].map(m3)

# Mappings from df2 to merged_df
m4 = df2.set_index("ISO")["Total firearms"]
merged_df["Total firearms"] = merged_df["ISO"].map(m4)

# Mappings from df3 to merged_df
m5 = df3.set_index("ISO")["Total military firearms"]
merged_df["Total military firearms"] = merged_df["ISO"].map(m5)

merged_df.rename(columns={'Total firearms': 'Total law enforcement firearms'}, inplace=True)

merged_df.head(10)
```

```python
merged_df.to_csv("merged_dataset.csv", index=False) 
```

# Adding GDP to merged dataset with homicide and calculating per capitas. Also adding violent & sexual crime numbers

```python
df = pd.read_csv('merged_homicide_weapon_no_NULL.csv')
desired_order = ["ISO", "Country", "Continent", "Subregion", "Year", "Population", "Estimate of firearms in civilian possession", 
                 "Total law enforcement firearms", "Total military firearms", "Total Homicides"]
df = df[desired_order]

gdp = pd.read_excel('./separate_datasets/GDP.xlsx')
sexual_violent_df = pd.read_excel('../crime/data_cts_violent_and_sexual_crime.xlsx')
```

```python
gdp.head(10)
```

```python
mapping1 = gdp.set_index("Country Code")[2018]
df["Country GDP"] = df["ISO"].map(mapping1)

df.head(10)
```

```python
columns_to_convert = [
    "Estimate of firearms in civilian possession",
    "Total law enforcement firearms",
    "Total military firearms",
    "Total Homicides",
    "Country GDP",
]

for column in columns_to_convert:
    df[f"{column} per 1,000"] = (df[column] / df["Population"]) * 1000

df.head(10)
```

```python
category_mapping = {
    "Sexual assault": "Sexual Crimes",
    "Sexual Exploitation": "Sexual Crimes",
    "Sexual violence": "Sexual Crimes",
    "Sexual violence: Other acts of sexual violence": "Sexual Crimes",
    "Sexual violence: Rape": "Sexual Crimes",
    "Sexual violence: Sexual assault": "Sexual Crimes",
    "Acts intended to induce fear or emotional distress": "Violent Crimes",
    "Acts intended to induce fear or emotional distress:": "Violent Crimes",
    "Intimate partner or family member": "Violent Crimes",
    "Kidnapping": "Violent Crimes",
    "Other Perpetrator known to the victim": "Violent Crimes",
    "Perpetrator unknown to the victim": "Violent Crimes",
    "Relationship to perpetrator is not known": "Violent Crimes",
    "Serious assault": "Violent Crimes",
}

sexual_violent_df['Main Category'] = sexual_violent_df['Category'].map(category_mapping)

result = sexual_violent_df.groupby(['ISO', 'Year', 'Main Category'])['VALUE'].sum().reset_index()
result = result.pivot(index=['ISO', 'Year'], columns='Main Category', values='VALUE').reset_index()

result.columns.name = None 
result = result.rename(columns={'Violent Crimes': 'Total Violent Crimes', 'Sexual Crimes': 'Total Sexual Crimes'})

result = result[result['Year'] == 2018]

result.head(20)
```

```python
mapping1 = result.set_index("ISO")["Total Sexual Crimes"]
df["Total Sexual Crimes"] = df["ISO"].map(mapping1)

mapping2 = result.set_index("ISO")["Total Violent Crimes"]
df["Total Violent Crimes"] = df["ISO"].map(mapping2)

df['Total Sexual Crimes'] = df['Total Sexual Crimes'].fillna(0).apply(round)
df['Total Violent Crimes'] = df['Total Violent Crimes'].fillna(0).apply(round)

columns_to_convert = [
    "Total Sexual Crimes",
    "Total Violent Crimes"
]

for column in columns_to_convert:
    df[f"{column} per 1,000"] = (df[column] / df["Population"]) * 1000

new_order = [
    "ISO",
    "Country",
    "Continent",
    "Subregion",
    "Year",
    "Population",
    "Estimate of firearms in civilian possession",
    "Total law enforcement firearms",
    "Total military firearms",
    "Estimate of firearms in civilian possession per 1,000",
    "Total law enforcement firearms per 1,000",
    "Total military firearms per 1,000",
    "Total Homicides",
    "Total Sexual Crimes",
    "Total Violent Crimes",
    "Total Homicides per 1,000",
    "Total Sexual Crimes per 1,000",
    "Total Violent Crimes per 1,000",
    "Country GDP",
    "Country GDP per 1,000"
]

df = df[new_order]

df.head(10)
```

```python
df.to_csv("everything_merged_dataset_with_NULL_violent&sexual.csv", index=False) 
```
