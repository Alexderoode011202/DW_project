# Law Enforcement Data Cleaning

```python
import pandas as pd
import numpy as np
```

### Re-assigning the first row as the list for column names

```python
dataset = pd.read_excel("SAS-BP-Law-enforcement-firearms-annexe.xlsx")
dataset.columns=dataset.iloc[0]
dataset.head()
```

### Here the computation method gets dropped, together with category

Both attributes are irrelevant for the research we are conducting here

```python
dataset = dataset.drop(["Category", "Computation method"], axis=1)
dataset
```

### Some colummns need to have their names changed due to conversion errors

```python
dataset.columns=["ISO", "Country", "Continent", "Subregion", "Population", "Unnamed 1", "Active personel", "Law Enforcement Firearms", "Unnamed 2"]
dataset
```

```python
dataset = dataset.drop("Unnamed 1", axis=1)
```

```python
columns = list(dataset.columns)
columns[-2] = "Total firearms"  # "total" is supposed to be "Total Firearms"
columns[-3] = "Active Personel" # "active" is supposed to be "Active Personel"
dataset.columns=columns
dataset
```
