# Civilian dataset cleaning

```python
import pandas as pd
import numpy as np
from typing import List
```

### set the correct column names:

```python
civilian_data = pd.read_excel("SAS-BP-Civilian-held-firearms-annexe.xlsx")
civilian_data.columns = civilian_data.iloc[0]
civilian_data.head()
```

```python
civilian_data = civilian_data.drop([0,1,2], axis=0)
civilian_data.head()
```

### move the values from column 3 to column 2 if need be

```python
civilian_data.index = range(0, civilian_data.shape[0])
for idx in range(0, civilian_data.shape[0]):
    if civilian_data.iloc[idx, 3] in ["America", "Europe", "Oceania", "Africa", "Asia"]:
        civilian_data.iloc[idx, 2] = civilian_data.iloc[idx, 3]
    
```

```python
civilian_data
```

### Rename the columns again and drop the unnamed/irrelevant ones

```python
civilian_data.columns = ["Country code", "Country", "Region", "unnamed 1", "Subregion", "Population 2017", "unnamed 2", "unnamed 3", "Estimate of firearms in civiian Posession","amount of weapons per 100 persons", "unnamed 4", "unnamed 5", "unnamed 6", "unnamed 7", "Registered firearms", "Unregistered firearms"]
civilian_data.drop(["unnamed 1", "unnamed 2", "unnamed 3", "unnamed 4", "unnamed 5", 'unnamed 6', "unnamed 7"], axis=1)
```

### And finally save the cleaned dataset

```python
civilian_data.to_csv("SAS-BP-Civilian-held-firearms-annexe.csv")
```
